```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('laptop_price.csv',encoding='latin1')
data
```

```python
data['Ram']=data['Ram'].str.replace('GB','').astype('int32')
```

```python
data.head()
```

```python
data['Weight']=data['Weight'].str.replace('kg','').astype('float32')
```

```python
data.head()
```

```python
data['Company'].value_counts()
```

```python
def add_company(inpt):
    if inpt=='Samsung' or inpt=='Mediacom' or inpt=='Razer' or inpt=='Microsoft' or inpt=='Vero' or inpt=='Xiaomi' or inpt=='Chuwi' or inpt=='Fujitsu' or inpt=='Google' or inpt=='LG' or inpt=='Huawei':
        return 'other'

    else:
        return inpt 
data['Company']=data['Company'].apply(add_company)
```

```python
data['Company'].value_counts()
```

```python
data['ScreenResolution'].value_counts()
```

```python
data['Touchscreen']=data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['IPS']=data['ScreenResolution'].apply(lambda x:1 if'IPS' in x else 0)
```

```python
data.head(1)
```

```python
data['Cpu'].value_counts()
```

```python
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
def set_processor(name):
    if name=='Intel Core i5' or name=='Intel Core i7' or name=='Intel Core i3':
        return name
    else:
        if name.split()[0]=='AMD':
            return 'AMD'

        else:
            return 'other'
        

data['cpu_name']=data['cpu_name'].apply(set_processor)
```

```python
data.head(3)
```

```python
data['Gpu'].value_counts()
```

```python
data['Gpu_name']=data['Gpu'].apply(lambda x:" " .join(x.split()[0:1]))
```

```python
data['Gpu_name'].value_counts()
```

```python
data=data[data['Gpu_name']!='ARM']
```

```python
data.shape
```

```python
data.head(2)
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(inpt):
    if inpt=='Windows 10' or inpt=='Windows 7' or inpt=='Windows 10 S':
        return 'Windows'

    elif inpt=='macOS' or inpt=='Mac OS X':
        return 'Mac'

    elif inpt=='Linux':
        return inpt
    else:
        return 'Other'

data['OpSys'] =data['OpSys'].apply(set_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.head(3)
```

```python
data=data.drop(columns=['laptop_ID','Inches','Product','ScreenResolution','Gpu','Cpu'])
```

```python
data
```

```python
data.shape
```

```python
data=pd.get_dummies(data)
```

```python
data
```

```python
x=data.drop('Price_euros',axis=1)
y=data['Price_euros']
```

```python
x.head(2)
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
def model_acc(model):
    model.fit(x_train, y_train)
    acc = model.score(x_train, y_train)
    print(f"{str(model)} -- {str(acc)}")
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
ls=Lasso()
model_acc(ls)

from sklearn.tree import DecisionTreeRegressor
ds=DecisionTreeRegressor()
model_acc(ds)

from sklearn.ensemble import RandomForestRegressor
rn=RandomForestRegressor()
model_acc(rn)
```

```python
from sklearn.model_selection import GridSearchCV
parameters={'n_estimators':[10,50,100],
            'criterion':['squared_error','absolute_error','poisson']}
grid_obj=GridSearchCV(estimator=rn,param_grid=parameters)

grid_fit=grid_obj.fit(x_train,y_train)

best_model=grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test,y_test)
```

```python
import pickle
with open('predictor.pickle','wb')as file:
    pickle.dump(best_model,file)
```
